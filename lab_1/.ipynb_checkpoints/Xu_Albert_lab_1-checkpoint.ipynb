```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from scipy.io.arff import loadarff 
```

# **2 studying integrated pulse profile**

```python
HTRU_data_csv = pd.read_csv("./htru2/HTRU_2.csv",header=None)
HTRU_data_arff = pd.DataFrame(loadarff("./htru2/HTRU_2.arff")[0])
#HTRU_data_csv.iloc[0]
HTRU_data_arff
#HTRU_data_arff.iloc[0]
#is there a reason the csv has unlabeled columns and the arff doesn't?
```

```python
data = HTRU_data_arff
noise = data.loc[data["class"] == b'0']
real = data.loc[data["class"] == b'1']

# min_mean, max_mean = min(data["Profile_mean"]), max(data["Profile_mean"])
# bins = []
# plot.hist(real["Profile_mean"], alpha=0.5, density=True)
# plot.hist(noise["Profile_mean"], alpha=0.5, density=True)
# plot.show()
```

```python
# 3
for index in range(4):
    min_feature, max_feature = min(data.iloc[:,index]), max(data.iloc[:,index])
    bins = np.arange(min_feature, max_feature, (max_feature-min_feature)/50)
    plot.hist(real.iloc[:,index], alpha=0.5, density=True,bins=bins,edgecolor='black')
    plot.hist(noise.iloc[:,index], alpha=0.5, density=True,bins=bins,edgecolor='black')
    plot.title(data.columns[index])
    plot.xlabel(data.columns[index])
    plot.ylabel("bin width (probability)")
    
    #show mean, std as text box instead of text
    fr_mn, fr_std = real.iloc[:,index].mean(), real.iloc[:,index].std()
    fn_mn, fn_std = noise.iloc[:,index].mean(), noise.iloc[:,index].std()
    print(f'real mean: {fr_mn} real std: {fr_std}')
    print(f'noise mean: {fn_mn} noise std: {fn_std}')
    
    plot.show()
```

```python
# 4
#probability(is pulsar candidate) = real/(real+noise)
P_real = len(real)/(len(real)+len(noise))
print(f'P_real= {P_real}')

#P(Mean<80)
mean_less_80 = (data['Profile_mean']<80).sum()
print(f'P(Mean<80)= {mean_less_80/len(data)}')

#P(both)
p_both = len(data[(data['Profile_mean']<80) & (data['class']==b'1')])/len(data)
print(f'P(both)= {p_both}')
```

```python
# 5
# P(A|B) = P(B|A)*P(A)/P(B)

# P(B) = P(Mean<80)
P_B = len(data[data['Profile_mean']<80])/len(data)

# P(A) = P(pulsar)
P_A = len(real)/len(data)

# P(B|A) = P(mean<80 given pulsar)
P_B_given_A = len(data[(data['Profile_mean']<80) & (data['class']==b'1')])/len(real)

# P(A|B) = P(pulsar given mean<80)
P_A_given_B = P_B_given_A*P_A/P_B
print(f'P(pulsar given mean<80)= {P_A_given_B}')
```
