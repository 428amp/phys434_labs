```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import scipy
```

```python
mu=3
sigma=.01
#50 bin histogram of 100k sampling of normal distribution mean 3, width .01
u_is = scipy.stats.norm.rvs(size=100000,loc=mu,scale=sigma)
plot.hist(u_is,density=True,edgecolor='black',alpha=.5,bins=50)

#gaussian pdf (y axis percentages)
M, step=np.linspace(scipy.stats.norm.ppf(0.0001, loc=mu, scale=sigma), scipy.stats.norm.ppf(0.9999, loc=mu, scale=sigma),num=100000,retstep=True)
plot.plot(M,scipy.stats.norm.pdf(M,loc=mu,scale=sigma))

plot.xlim(2.95, 3.05)
plot.xlabel("var")
plot.ylabel("probability(%)")
plot.title("hist vs pdf")
plot.show()
```

```python
plot.hist(u_is,density=True,edgecolor='black',alpha=.5,bins=50)
plot.plot(M,scipy.stats.norm.pdf(M,loc=mu,scale=sigma))
plot.xlim(2.95, 3.05)
plot.xlabel("var")
plot.ylabel("probability(%)")
plot.title("hist vs pdf (log)")
plot.yscale('log')
plot.show()
```
