```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
N=1000
fig, ax = plt.subplots()
points = np.random.uniform(0, 1, size=(N,2))
ax.scatter(points[:,0],points[:,1],s=.5)
circle1 = plt.Circle((0, 0), 1, color='r', alpha=.5)
ax.add_patch(circle1)
ax.set_aspect('equal')
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()
```

```python
#pi estimation
#count num inside, which is num with x^2+y^2<1

valid = 0
for i in range(len(points)):
    x = points[i,0]
    y = points[i,1]
    # print(f'{x},{y}')
    if x**2+y**2 < 1:
        valid+=1
print(f'valid points/1000: {valid}')

# pi estimate logic: quarter circle has area pi/4, so multiply valid/1000 by 4
pi_estimate = valid*4/1000
print(f'pi estimate:{pi_estimate}')
```

```python
#try again w/ N=1e4, 1e6
def pi_estimate_N(N):
    print(f'estimate pi for sample N={N}')
    fig, ax = plt.subplots()
    points = np.random.uniform(0, 1, size=(N,2))
    ax.scatter(points[:,0],points[:,1],s=.5)
    circle1 = plt.Circle((0, 0), 1, color='r', alpha=.5)
    ax.add_patch(circle1)
    ax.set_aspect('equal')
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.show()

    valid = 0
    for i in range(len(points)):
        x = points[i,0]
        y = points[i,1]
        # print(f'{x},{y}')
        if x**2+y**2 < 1:
            valid+=1
    print(f'valid points/{N}: {valid}')

    pi_estimate = valid*4/N
    print(f'pi estimate:{pi_estimate}')
    print('____')

pi_estimate_N(int(1e4))
pi_estimate_N(int(1e6))
```
