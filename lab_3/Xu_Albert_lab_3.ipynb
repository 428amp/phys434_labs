```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from scipy.stats import beta
import scipy
```

```python
x = np.linspace(0, 1, 1002)[1:-1]
fig, ax = plot.subplots(figsize=(5, 3.75))
a=5.0
b=3
dist = beta(a, b)
plot.plot(x, dist.pdf(x), ls="-", c='black',label=r'$\alpha=%.1f,\ \beta=%.1f$' % (a, b))
plot.xlim(0, 1)
plot.ylim(0, 3)
plot.xlabel('$x$')
plot.ylabel(r'$p(x|\alpha,\beta)$')
plot.title('Beta Distribution')
plot.legend(loc=0)
plot.show()
```
