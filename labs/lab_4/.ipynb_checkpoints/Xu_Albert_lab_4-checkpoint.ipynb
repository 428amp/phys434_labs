```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from scipy.stats import beta
import scipy
```

```python
#in general, fix your data structures to be more easily usable + efficient
# read numpy docs!

mass = [
7.95442531,
10.60408232,
9.9020277,
10.42288788,
17.2193401,
10.06076572,
10.00176003,
7.8289388,
9.55481385,
9.99981543,
10.3156156,
10.70547087,
9.80374407,
7.62085689,
9.60747351,
]
```

## **Task 1: visualization**

unbinned mass distribution:

```python
plot.scatter(mass, np.zeros(len(mass)), s=5)
plot.title("unbinned mass distribution")
plot.xlabel("GeV")
```

binned mass distribution w/ error bar height $\propto\sqrt{bin\ height}$

```python
n, bins, patches = plot.hist(mass,bins=np.arange(0,20,.5),edgecolor='black',alpha=.5)
errors = np.sqrt(n)
bin_centers = 0.5 * (bins[1:] + bins[:-1])
plot.errorbar(bin_centers, n, yerr=errors, fmt='none', capsize=3,color="red")
plot.title("mass distribution, .5 GeV bins")
plot.xlabel("GeV")
plot.ylabel("count")
plot.show()
```

## **Task 2: Maximum likelihood estimate**

#### **1.** unbinned negative log-likelihood function

```python
def NLL(m, gamma):
    return -2*np.sum(np.log(scipy.stats.cauchy.pdf(mass,loc=m,scale=gamma)))

# scan over mass, width (m, gamma)
# linspace vs arange inclusive endpoints etc. ? 
# https://stackoverflow.com/questions/62106028/what-is-the-difference-between-np-linspace-and-np-arange
m_vals = np.linspace(0, 20, num=100)
gamma_vals = np.linspace(1e-5, 10, num=250)
# jank right now to turn this into contour plot
NLL_vals = [[NLL(m, gamma) for m in m_vals] for gamma in gamma_vals]
NLL_vals = np.array(NLL_vals)

fig, ax = plot.subplots()
levels = np.arange(50, 200, 12)
CS = ax.contour(m_vals, gamma_vals, NLL_vals,levels=levels)
ax.clabel(CS, inline=True, fontsize=10)

plot.xlim(0,20)
plot.ylim(0,2)

plot.xlabel("m")
plot.ylabel("gamma")
plot.title("NLL contour plot over gamma, m")
plot.show()
```

```python
i, j = np.where(NLL_vals == np.min(NLL_vals))
m=m_vals[j]
gamma=gamma_vals[i]
plot.hist(mass,density=True,bins=np.arange(0,20,.5),edgecolor='black',alpha=.5)
M=np.linspace(scipy.stats.cauchy.ppf(0.001, loc=m, scale=gamma), 20, 500)
plot.plot(M,scipy.stats.cauchy.pdf(M,loc=m,scale=gamma))
plot.xlim(0,20)
plot.title('best fit PDF superposed with mass distribution')
plot.xlabel('GeV')
plot.ylabel('probability')
plot.show()
```
