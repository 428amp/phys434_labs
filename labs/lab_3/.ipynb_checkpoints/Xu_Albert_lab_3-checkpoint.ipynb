```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from scipy.stats import beta
import scipy
```

## **1: rejection method**

example beta distribution for alpha=5, beta=3

```python
x = np.linspace(0, 1, 1002)[1:-1]
fig, ax = plot.subplots(figsize=(5, 3.75))
a=5.0
b=3
dist = beta(a, b)
plot.plot(x, dist.pdf(x), ls="-", c='black',label=r'$\alpha=%.1f,\ \beta=%.1f$' % (a, b))
plot.xlim(0, 1)
plot.ylim(0, 3)
plot.xlabel('$x$')
plot.ylabel(r'$p(x|\alpha,\beta)$')
plot.title('Beta Distribution')
plot.legend(loc=0)
plot.show()
```

maximize beta function by setting derivative to 0, giving us max at x=(a-1)/(a+b-2)

```python
x=(a-1)/(a+b-2)
beta_max = scipy.special.gamma(a+b)*x**(a-1)*(1-x)**(b-1)*1/scipy.special.gamma(a)*1/scipy.special.gamma(b)
print(beta_max)
```

value of beta_max passes smell test by inspecting graph

for event weight formula, we use the PDF for beta

```python
#1.1
def f_pdf(x, a, b):
    return scipy.special.gamma(a+b)*x**(a-1)*(1-x)**(b-1)*1/scipy.special.gamma(a)*1/scipy.special.gamma(b)
    
accepted = np.zeros(1000)

total_accepted = 0
total_iter = 0
while total_accepted < 1000:
    x = scipy.stats.uniform.rvs()
    y = scipy.stats.uniform.rvs()
    total_iter += 1
    if f_pdf(x,5,3) < y:
        # accepted[total_accepted] = x
        total_accepted += 1

print(f'iters needed to accept 1000: {total_iter}')
```

we'd expect total iters needed to accept 1000 to be on the order of beta_max*1000

## **2: sphere in 3d**

```python
# 2.1
def sphere_index_function(e,R):
    return e[0]**2+e[1]**2+e[2]**2<R**2

# N = 100
N_vals, m_vals = [], []
rho = 2
for N in range(100, 100000, 1000):
    N_points = np.random.rand(N,3)*2-1
    # m = 2**3/N*np.sum(N_points)
    V=8
    m=0
    for e in N_points:
        m += rho*sphere_index_function(e,1)*V/N
    N_vals.append(N)
    m_vals.append(m)
plot.plot(N_vals,m_vals)
plot.ylim(5, 10)
plot.xlabel('N')
plot.ylabel('mass')
plot.title('mass vs N')
plot.show()
# fix some bits here (like implementation for speed, etc.)
```

mass quickly converges on actual value of $\frac{4}{3}\pi\rho r^3$ as we crank N up
