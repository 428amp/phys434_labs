```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from scipy.io.arff import loadarff 
```

#### **1.** Download dataset.
#### **2.** Choose features for study.
##### studying features from integrated pulse profile.

```python
HTRU_data_csv = pd.read_csv("./htru2/HTRU_2.csv",header=None)
HTRU_data_arff = pd.DataFrame(loadarff("./htru2/HTRU_2.arff")[0])

# compare csv vs arff after load
HTRU_data_csv.iloc[0]
HTRU_data_arff.iloc[0]

#is there a reason the csv has unlabeled columns and the arff doesn't? 
#is it just because a csv can't support column labels if it's literally just comma separate values?
#regardless, because of this choosing to use arff as it makes unmodified dataframe a little easier to work with

HTRU_data_arff
```

#### **3.** Compare shape of features between classes.
##### noise, real separated by examining for each row whether class is 1(real) or 0(noise)

```python
data = HTRU_data_arff
noise = data.loc[data["class"] == b'0']
real = data.loc[data["class"] == b'1']

# figuring out visualization, not used
# min_mean, max_mean = min(data["Profile_mean"]), max(data["Profile_mean"])
# bins = []
# plot.hist(real["Profile_mean"], alpha=0.5, density=True)
# plot.hist(noise["Profile_mean"], alpha=0.5, density=True)
# plot.show()
```

##### for each feature, separate all data (both noise+real) into ~50 bins and plot normalized histograms for real(blue-), noise(orange--)

```python
for index in range(4):
    min_feature, max_feature = min(data.iloc[:,index]), max(data.iloc[:,index])
    binwidth=(max_feature-min_feature)/50
    bins = np.arange(min_feature, max_feature, binwidth)
    plot.hist(real.iloc[:,index], alpha=0.5, density=True,bins=bins,edgecolor='black')
    plot.hist(noise.iloc[:,index], alpha=0.5, density=True,bins=bins,edgecolor='black',linestyle='--')
    plot.title(data.columns[index],fontweight="bold")
    plot.xlabel(data.columns[index])
    plot.ylabel(f"probability (binwidth={round(binwidth,2)})")
    
    fr_mn, fr_std = real.iloc[:,index].mean(), real.iloc[:,index].std()
    fr_mn, fr_std = round(fr_mn, 2), round(fr_std, 2)
    fn_mn, fn_std = noise.iloc[:,index].mean(), noise.iloc[:,index].std()
    fn_mn, fn_std = round(fn_mn, 2), round(fn_std, 2)
    
    # show mean, std as text
    # print(f'real mean: {fr_mn} real std: {fr_std}')
    # print(f'noise mean: {fn_mn} noise std: {fn_std}')

    # text box to dump mean, std to 2 decimal points
    # is there a less annoying way to do this?
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax = plot.gca()
    box_contents = f'real mean: {fr_mn} real std: {fr_std}\nnoise mean: {fn_mn} noise std: {fn_std}'
    plot.text(0.02,.97, box_contents, ha='left', va='top', transform=ax.transAxes, bbox=props)
    
    plot.show()
```

#### **4.** Compute some probabilities.
##### in general $P(A) = \frac{count\_A}{count\_total}$

```python
#P(is_pulsar)
P_real = len(real)/(len(real)+len(noise))
print(f'P_real= {P_real}')

#P(Mean<80)
mean_less_80 = (data['Profile_mean']<80).sum()
print(f'P(Mean<80)= {mean_less_80/len(data)}')

#P(both)
p_both = len(data[(data['Profile_mean']<80) & (data['class']==b'1')])/len(data)
print(f'P(both)= {p_both}')
```

#### **5.** Compute some conditional probabilities using Bayes' theorem.
##### Bayes' theorem: $P(A|B) = P(B|A)*P(A)/P(B)$

```python
# P(A|B) = P(B|A)*P(A)/P(B)

# P(B) = P(Mean<80)
P_B = len(data[data['Profile_mean']<80])/len(data)

# P(A) = P(pulsar)
P_A = len(real)/len(data)

# P(B|A) = P(mean<80 given pulsar)
P_B_given_A = len(data[(data['Profile_mean']<80) & (data['class']==b'1')])/len(real)

# P(A|B) = P(pulsar given mean<80)
P_A_given_B = P_B_given_A*P_A/P_B
print(f'P(pulsar given mean<80)= {P_A_given_B}')
```
